```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score

from process import load_dataset, METRICS_CONFIGS
sns.set_theme()
```

```python
metrics_config_index = 0
```

```python
train_dataset, test_dataset = load_dataset(metrics_config_index, True), load_dataset(
    metrics_config_index, False
)
```

```python
train_dataset
```

```python
test_dataset
```

```python
from sklearn.preprocessing import StandardScaler

x_cols = ["DURATION"] + METRICS_CONFIGS[metrics_config_index]
train_X, train_Y = train_dataset[x_cols].to_numpy(), train_dataset["CLASS"].to_numpy()
test_X, test_Y = test_dataset[x_cols].to_numpy(), test_dataset["CLASS"].to_numpy()
```

```python
scaler = StandardScaler()
scaler.fit(train_X)
train_X = scaler.transform(train_X)
test_X = scaler.transform(test_X)
```

```python
# Train and test an SVM
from sklearn.svm import SVC

clf = SVC()
clf.fit(train_X, train_Y)
pred_Y = clf.predict(test_X)
accuracy_score(test_Y, pred_Y)
```
